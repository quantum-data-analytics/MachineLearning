## Stochastic Gradient Descent (SGD) Explained with Python Code

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data

X = 2 * np.random.rand(100,1)
y = 2* X + 1 + np.random.randn(100,1)
```

```python
# Visualize the data

plt.scatter(X,y)
plt.xlabel('X')
plt.ylabel('y')
plt.title("Sample Data")
plt.show()
```

```python
# Implementation of SGD Linear Equation

theta = np.random.randn(2,1) # Random Initalization of theta ( theta0, theta1)
learning_rate = 0.1
n_epochs = 100
m = len(X)

# Add bias tern to X
X_b = np.c_[np.ones((m,1)),X]

# Stochastic GRadient Descent (SGD)
for epoch in range(n_epochs):
    for i in range(m):
    # Randomly pick a single sample
        random_index = np.random.randint(m)
        xi = X_b[random_index: random_index +1]
        yi = y[random_index: random_index+1]
    
        # compute the prediction
        prediction = xi.dot(theta)
    
        # compute the gradient
        gradient = 2 * xi.T.dot(prediction - yi)
    
        # update the parameter
        theta = theta - learning_rate * gradient

# Display the learned parameters
print("learned Parameter", theta)

```

```python
# Visualize the model learned rate
# Visualize the model's learned line
plt.scatter(X, y, label='Data')
plt.plot(X, X_b.dot(theta), color='red', label='Fitted line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression with SGD')
plt.show()
```
