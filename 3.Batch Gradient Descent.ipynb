## Batch Gradient Descent

Gradient Descent is an optimization algorithm used to minimize a function i.e. cost function or loss function - by iteratively adjustin the parameter of the model.

Formula: 
θ = θ- α.▽J(θ)
θ =  the parameter we are trying to optimize 
α = the learning rate
▽J(θ) = gradient of the cost function

```python
# compute cost
def compute_cost(X,y, theta):
    m = len(y)
    predictions = X.dot(theta) 
    cost = (1/(2*m))*np.sum(np.square(predictions-y)) # MSE Formula
    return cost
```

```python
# gradient descent

def gradient_descent(X,y, theta, alpha, num_iterations):
    m = len(y)
    cost_history = []

    for i in range(num_iterations):
        gradients = (1/m) * X.T.dot(X.dot(theta)-y) # Gradeint of MSE
        theta = theta - alpha * gradients # update the parameter
        cost_history.append(compute_cost(X,y, theta)) #Track cost function 
    return theta, cost_history
```

```python
# Gradient Descent implemenatation
X_b = np.c_[np.ones((X.shape[0],1)),X] # Adding bias term (X0=1)
theta_initial = np.random.randn(2,1) # Random initial parametes 
alpha = 0.01 # Learning rate
num_iteration = 100

# Run gradient descent 
theta_optimal, cost_history = gradient_descent(X_b,y, theta_initial, alpha, num_iteration)

print("Optimal paramters:", theta_optimal)
print("Final cost:", cost_history[-1])
```

```python
# plotting the cost history
plt.plot(range(num_iteration), cost_history)
plt.xlabel("Number of iteration")
plt.ylabel("Cost (MSE)")
plt.title("Cost Function History")
plt.show()
```
