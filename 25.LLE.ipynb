## Locally Linear Embedded

## What is LLE?
- Locally Linear embedded is a non-linear dimenionality reduction dimensionality reduction technique.

## Why use LLE?
- Works well with non-linear data
- Better for capturing local relationship
- Useful for high-dimensional data visualization. 

```python
# Libraries 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import LocallyLinearEmbedding
```

```python
# Generate data
X, color = make_swiss_roll(n_samples = 1500, noise = 0.1)

# Visualize the data in 3D

fig = plt.figure(figsize = (10, 8))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c = color, cmap = plt.cm.Spectral)
plt.show()
```

```python
# Initialize LLE
lle = LocallyLinearEmbedding(n_neighbors=12, n_components=2, method='standard')
```

```python
# Apply LLE
X_lle = lle.fit_transform(X)
```

```python

# Visualize the 2D embeddig of the data

plt.figure(figsize=(10, 7))
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=color, cmap=plt.cm.Spectral)
plt.title("Swiss Roll Data after LLE")
plt.show()
```

## Tuning parameters

```python
# Adjusting parameters
lle = LocallyLinearEmbedding(n_neighbors=30, n_components=2, method='hessian')

# Applying LLE to the Swiss Roll data again
X_lle = lle.fit_transform(X)

# Visualize the modified LLE result
plt.figure(figsize=(8, 6))
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=color, cmap=plt.cm.Spectral)
plt.title("Modified Swiss Roll Data After LLE")
plt.show()
```
