## SVM Regression

# What is SVM Regression?

SVM Regression aims to find a function that approximates the relationship between input features and continous output values. 

# Important paramter 
1. C (Regularization parameter)
2. Epsilon

# Chosing the Kernel
1. Linear 
2. Polynomial Kernel
3. Radical Basis Function

```python
# Libraries 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples = 100, n_features=1, noise = 30, random_state=42)
```

```python
# Visualize the data
plt.scatter(X, y, color='blue')
plt.title("Data for SVM Regression")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.show()
```

```python
# SVM Regression Model 
from sklearn.svm import SVR

svr = SVR(kernel='linear',  C = 100, epsilon=0.1)

# fit

svr.fit(X,y)

# predict 
y_pred = svr.predict(X)
```

```python
# Plot the results
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='SVR model')
plt.title("SVM Regression with Linear Kernel")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()
plt.show()
```

```python

# Create an SVR model with RBF kernel
svr_rbf = SVR(kernel='rbf', C=100, epsilon=0.1)

# Fit the model to the data
svr_rbf.fit(X, y)

# Predict using the trained model
y_pred_rbf = svr_rbf.predict(X)

# Plot the results
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred_rbf, color='green', label='RBF SVR model')
plt.title("SVM Regression with RBF Kernel")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()
plt.show()
```

```python
# Evaluation

from sklearn.metrics import mean_squared_error

mse_rdf = mean_squared_error(y,y_pred_rbf)
mse_rdf
```
