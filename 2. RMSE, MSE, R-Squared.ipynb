## Regression Metrices

1. MSE: Mean Squared Error: It measures the average squared difference between the predicted and actual values. The smaller the value, the better the model fits. 
2. RMSE: Root Mean Squared Error: Its just the square root of MSE. 
3. R-Squared: This givs us the proposition of variances in the dependednt variables that is predictable from the independent variables. A higher R-squared value indcates a better fit. 

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# sample data
np.random.seed(42)
x = np.random.rand(100,1)*10  #Random feature
y = 3*x + np.random.randn(100,1) *2 # Linear Relation with noise
```

```python
# split data into train and test
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)

# make predictions based on the test
y_pred = model.predict(X_test)
```

```python
# Calcluate MSE
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error:{mse}")
```

```python
# RMSE
rmse = np.sqrt(mse)
print(f"RMSE is {rmse}")
```

```python
# Calcuate R-Squared 
r2 = r2_score(y_test, y_pred)
print(f"R-Squared is {r2}")
```

```python
import matplotlib.pyplot as plt

# plot true vs predicted values 
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color = 'red', lw = 2)
plt.xlabel("True Values")
plt.ylabel("Predicted Values")
plt.title("True vs Predicted Values")
plt.show()
```
