## Linear Regression Explained

### What is Linear Regression?

#### Linear Regression is a method used to model the relationship between the dependent variable and one or more independent variables  by fitting a linear equation to observed data

# Equation
y = mx + b 
y = dependent variable 
x = independent variable 
m = slope
b= y-intercept


# Mean Squared Error


## Example

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate random data 
np.random.seed(42)
heights = np.random.normal(160, 10, 100) # mean # std
weights = 0.5 * heights + np.random.normal(0,5,100)
```

```python
heights
```

```python
weights
```

```python
plt.scatter(heights, weights, color = 'blue')
plt.title("Heights vs Weights")
plt.xlabel("Heights (cm)")
plt.ylabel("Weight (kg")
plt.show()
```

```python
heights = heights.reshape(-1,1)
weights = weights.reshape(-1,1)
```

```python
heights
```

```python
# create and fit the model
model = LinearRegression()
model.fit(heights, weights)
```

```python
plt.scatter(heights, weights, color = 'blue')
plt.plot(heights, model.predict(heights), color = 'red', linewidth = 2)
plt.title("Linear Regression: Height vs Weight")
plt.xlabel("Heights (cm)")
plt.ylabel("Weight (kg")
plt.show()
```

```python
slope = model.coef_[0]
intercept = model.intercept_
slope, intercept
```

```python
# R-Squared value
r_squared = model.score(heights, weights)
r_squared
```
